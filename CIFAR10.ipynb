```python
#importar datasets
from tensorflow import keras
import numpy as np
import tensorflow as tf
```

```python
## Cargar el conjunto de datos CIFAR-10
keras.datasets.cifar10.load_data()
```

```python
# Verificar las formas de las matrices
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
assert x_train.shape == (50000, 32, 32, 3)
assert x_test.shape == (10000, 32, 32, 3)
assert y_train.shape == (50000, 1)
assert y_test.shape == (10000, 1)
```

```python
# Verificar los rangos y tipos de datos
assert x_train.dtype == np.uint8, "El tipo de x_train no es uint8"
assert x_test.dtype == np.uint8, "El tipo de x_test no es uint8"
assert np.all(x_train >= 0) and np.all(x_train <= 255), "x_train contiene valores fuera del rango 0-255"
assert np.all(x_test >= 0) and np.all(x_test <= 255), "x_test contiene valores fuera del rango 0-255"
```
